```python
from crawler import Crawler

crawler = Crawler()
crawler.search("https://en.wikipedia.org/wiki/Internet", 1000,
               lambda soup: soup.find("h2", id="See_also"),
               lambda sibling: getattr(sibling.next_element, "name", None) == "h2")
```

```python
for el in crawler.getVisited():
  print(el.split("/")[-1])
```

```python
print("Iterations:",crawler.maxIterations())
print("Max depth:", crawler.maxDepth())
```
