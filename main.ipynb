```python
from crawler import Crawler

keywords = []
def findKeyword(url, data):
  if "internet" in data.lower():
    keywords.append(url)

crawler = Crawler()
crawler.search("https://en.wikipedia.org/wiki/Internet", 1000,
               lambda soup: soup.find("h2", id="See_also"),
               lambda sibling: getattr(sibling.next_element, "name", None) == "h2",
               lambda url, data: findKeyword(url, data))

print("this contains keyword `internet`")
for url in keywords:
  print(url)
```
